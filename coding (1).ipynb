```python
import pandas as pd
import numpy  as np
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

one_enc=OneHotEncoder(sparse_output=False)

cat_columns=df[['Geography','Gender']]

# geography_sparsed=one_enc.fit_transform(df['Geography'])
```

```python
cat_columns
```

```python
cat_sparsed=one_enc.fit_transform(cat_columns)
cat_sparsed
```

```python
import pandas as pd

cat_encoded=pd.DataFrame(cat_sparsed,columns=one_enc.get_feature_names_out())
cat_encoded
```

```python
df=pd.concat([df,cat_encoded],axis=1)
```

```python
df
```

```python
df=df.drop(['Geography','Gender'],axis=1)
```

```python
df
```

```python
X=df.drop('Exited',axis=1)
Y=df['Exited']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)

```

```python
X_test_scaled
```

```python
X_test_scaled
```

```python
X_train=pd.DataFrame(X_train_scaled,columns=X_train.columns,index=X_train.index)
```

```python
X_train
```

```python
X_test=pd.DataFrame(X_test_scaled,columns=X_test.columns,index=X_test.index)
```

```python
X_test
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers   import Dense
```

```python
model=Sequential(
    [
        Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
        Dense(32,activation='relu'),
        Dense(1,activation='sigmoid')
    ]
)
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
ls=tensorflow.keras.losses.BinaryCrossentropy()

model.compile(optimizer=opt,loss=ls,metrics=['accuracy'])

```

```python
history=model.fit(
    X_train,Y_train,validation_data=(X_test,Y_test),epochs=100
)
```

```python
model.save('model.h5')
```

```python
import pickle

with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)

with open('one_hot_encoder.pkl','wb') as file:
    pickle.dump(one_enc,file)
```
